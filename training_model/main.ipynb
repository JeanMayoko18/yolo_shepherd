```python
import torch
import cv2
import random
import pandas as pd
from pathlib import Path
from ultralytics import YOLO
from utils import get_all_available_gpus  
from tqdm import tqdm
import time
import gc
import os
import matplotlib.pyplot as plt

# -------------------------------
# Load images from a directory
# -------------------------------
def load_all_images(image_dir, max_images=-1):
    """
    Load up to max_images .jpg images from the specified directory.
    If max_images < 0, all images are loaded.

    Parameters:
    - image_dir (str): Path to the image directory.
    - max_images (int): Maximum number of images to load. If < 0, load all.

    Returns:
    - List of loaded images (as numpy arrays).
    """
    image_paths = list(Path(image_dir).glob("*.jpg"))
    image_paths = [p for p in image_paths if p.is_file()]  # Ensure valid paths

    if max_images > 0:
        image_paths = random.sample(image_paths, min(max_images, len(image_paths)))

    images = []
    for p in image_paths:
        img = cv2.imread(str(p))
        if img is not None:
            images.append(img)
        else:
            print(f"‚ö†Ô∏è Failed to read image: {p}")

    print(f"‚úÖ Loaded {len(images)} images from {image_dir}")
    return images

def load_images_(image_dir, max_images=-1):
    image_paths = sorted(Path(image_dir).glob("*.jpg"))
    image_paths = [p for p in image_paths if p.is_file()]
    if max_images > 0:
        image_paths = random.sample(image_paths, min(max_images, len(image_paths)))
    images = [cv2.imread(str(p)) for p in image_paths]
    return images, image_paths

def load_images(image_dir, max_images=-1, extensions=(".jpg", ".jpeg", ".png"), seed=42):
    """
    Load a fixed subset of images from a directory (deterministic sampling).

    Args:
        image_dir (str or Path): Directory containing images.
        max_images (int): Max number of images to load (-1 for all).
        extensions (tuple): Accepted file extensions.
        seed (int): Random seed for reproducible selection.

    Returns:
        images (List[np.ndarray]): Loaded images.
        image_paths (List[Path]): Corresponding file paths.
    """
    try:
        image_dir = Path(image_dir)
        image_paths = sorted(p for ext in extensions for p in image_dir.glob(f"*{ext}") if p.is_file())

        if max_images > 0 and len(image_paths) > max_images:
            random.seed(seed)
            image_paths = random.sample(image_paths, max_images)

        images = []
        valid_paths = []
        for p in image_paths:
            img = cv2.imread(str(p))
            if img is not None:
                images.append(img)
                valid_paths.append(p)
            else:
                print(f"‚ö†Ô∏è Unable to read image: {p}")
    except Exception as e:
        print(f"Error on loading images due to {e}" )
    return images, valid_paths

# -------------------------------
# Utility: model size and parameters
# -------------------------------
def get_model_size(model_path):
    """
    Get the size of a model file in megabytes.
    """
    return round(os.path.getsize(model_path) / (1024 * 1024), 2)

def count_model_params(model):
    """
    Count the total number of learnable parameters in the model.
    """
    return sum(p.numel() for p in model.model.parameters())

# -------------------------------
# Compute validation metrics
# -------------------------------
def get_model_metrics_global(metrics):
    precision = sum(metrics.p) / len(metrics.p) if metrics.p else None
    recall = sum(metrics.r) / len(metrics.r) if metrics.r else None
    f1 = sum(metrics.f1) / len(metrics.f1) if metrics.f1 else None
    map50 = metrics.map50 if hasattr(metrics, 'map50') else None

    return {
        "mAP@0.5": round(map50, 3) if map50 is not None else None,
        "Precision": round(precision, 3) if precision is not None else None,
        "Recall": round(recall, 3) if recall is not None else None,
        "F1 score": round(f1, 3) if f1 is not None else None
    }

def get_model_metrics_global2(metrics):
    precision = metrics.p[-1] if metrics.p else None
    recall = metrics.r[-1] if metrics.r else None
    f1 = metrics.f1[-1] if metrics.f1 else None
    map50 = metrics.map50 if hasattr(metrics, 'map50') else None

    return {
        "mAP@0.5": round(map50, 3) if map50 is not None else None,
        "Precision": round(precision, 3) if precision is not None else None,
        "Recall": round(recall, 3) if recall is not None else None,
        "F1 score": round(f1, 3) if f1 is not None else None
    }

def get_model_metrics(model, device, split="test", data_paths=None):
    try:
        val_results = model.val(device=device, split=split, save=True)
        metrics = getattr(val_results, 'box', None)
        if metrics is None:
            print("Warning: 'box' attribute not found in validation results")
            return {"mAP@0.5": None, "Precision": None, "Recall": None, "F1 score": None}

        precision = float(metrics.p[-1]) if metrics.p is not None and len(metrics.p) > 0 else None
        recall = float(metrics.r[-1]) if metrics.r is not None and len(metrics.r) > 0 else None
        f1 = float(metrics.f1[-1]) if metrics.f1 is not None and len(metrics.f1) > 0 else None
        map50 = float(metrics.map50) if hasattr(metrics, 'map50') else None

        f1_calc = 2 * precision * recall / (precision + recall) if precision and recall and (precision + recall) > 0 else None

        return {
            "mAP@0.5": round(map50, 3) if map50 is not None else None,
            "Precision": round(precision, 3) if precision is not None else None,
            "Recall": round(recall, 3) if recall is not None else None,
            "F1 score": round(f1_calc, 3) if f1_calc is not None else None
        }
    except Exception as e:
        print(f"Error_Metrics: Could not retrieve metrics due to error: {e}")
        return {"mAP@0.5": None, "Precision": None, "Recall": None, "F1 score": None}

# -------------------------------
# Inference benchmarking
# -------------------------------
def evaluate_model(
    name,
    model_path,
    images,
    gpu_idx,
    gpu_name,
    predict_mode=True,
    save_dir_root="results_predicted",
    yaml_path=None
) -> dict:
    """
    Evaluate YOLO model performance on selected images and GPU.

    Parameters:
    - name (str): Model name label.
    - model_path (str): Path to YOLO .pt model.
    - images (list): Image paths or arrays for inference.
    - gpu_idx (int): Index of target GPU.
    - gpu_name (str): GPU label.
    - predict_mode (bool): If True, uses model.predict(); else uses validation metrics.
    - save_dir_root (str): Root dir for saving predictions (used only in predict_mode).
    - yaml_path (str): Dataset YAML path (required if predict_mode=False).

    Returns:
    - dict with benchmark results and metrics.
    """
    device_str = f"cuda:{gpu_idx}"
    result = {
        "GPU": gpu_name,
        "Model": name,
        "Images processed": 0,
        "Total time (s)": 0,
        "Average FPS": 0,
        "Model size (MB)": 0,
        "Parameters": 0,
        "mAP@0.5": None,
        "Precision": None,
        "Recall": None,
        "F1 score": None,
        "Error": "",
        "InferenceMode": "predict" if predict_mode else "val/test"
    }
    print(len(images))
    try:
        model = YOLO(model_path)
        model.to(device_str)

        # Static model characteristics
        result["Model size (MB)"] = get_model_size(model_path)
        result["Parameters"] = count_model_params(model)

        # Warm-up
        _ = model.predict(images[0], device=device_str, verbose=False)

        # Predict mode
        if predict_mode:
            if not images:
                raise ValueError("No images provided for prediction.")

            save_dir = Path(save_dir_root) / name
            save_dir.mkdir(parents=True, exist_ok=True)

            start = time.time()
            for img in images:
                _ = model.predict(
                    img,
                    device=device_str,
                    verbose=False,
                    save=True,
                    save_dir=str(save_dir),
                    project=None,
                    stream=False
                )
            end = time.time()

        # Validation mode
        else:
            if not yaml_path:
                raise ValueError("YAML path is required when predict_mode=False.")

            metrics = {
                "mAP@0.5": None,
                "Precision": None,
                "Recall": None,
                "F1 score": None
            }

            try:
                
                raw_metrics = get_model_metrics(model, device=device_str, split="val")
                #print(f"[üß™ Debug] Raw output from get_model_metrics: {raw_metrics} (type: {type(raw_metrics)})")

                if isinstance(raw_metrics, tuple):
                    if len(raw_metrics) == 1 and isinstance(raw_metrics[0], dict):
                        raw_metrics = raw_metrics[0]
                    else:
                        raise ValueError(f"Unexpected tuple structure from get_model_metrics: {raw_metrics}")

                if not isinstance(raw_metrics, dict):
                    raise ValueError(f"Expected a dict from get_model_metrics but got {type(raw_metrics)}")

                for key in metrics:
                    value = raw_metrics.get(key)
                    if value is not None:
                        metrics[key] = value

            except Exception as e:
                print(f"[‚ùå Error in metric retrieval] {e}")
                        
            except Exception as metric_error:
                print(f"[Warning] Could not retrieve metrics for {name} on {gpu_name} due to: {metric_error}")

            # Optional timing of predictions for fairness
            start = time.time()
            for img in images:
                _ = model.predict(img, device=device_str, verbose=False)
            end = time.time()

            # Set metrics (even if some are None)
            result.update(metrics)
        # Common performance stats
        duration = end - start
        result.update({
            "Images processed": len(images),
            "Total time (s)": round(duration, 2),
            "Average FPS": round(len(images) / duration, 2)
        })

    except Exception as e:
        print(f"Error: Could not retrieve metrics due to error: {e}")
        result["Error"] = str(e)

    finally:
        if "model" in locals():
            del model
        gc.collect()
        torch.cuda.empty_cache()

    return result
#--------------------------------
# Graphical Comparison
#--------------------------------
def plot_comparison(df):
    import matplotlib.pyplot as plt
    import pandas as pd
    import numpy as np
    import textwrap

    required_columns = {"Model", "GPU", "Error"}
    if not required_columns.issubset(df.columns):
        missing = required_columns - set(df.columns)
        print(f"Missing required columns: {missing}")
        return

    df["Error"] = df["Error"].replace("", pd.NA)
    df_no_errors = df[df["Error"].isna()]
    if df_no_errors.empty:
        print("No valid data without errors to plot.")
        return

    metrics = [
        "Average FPS", "Model size (MB)", "Parameters",
        "mAP@0.5", "Precision", "Recall", "F1 score"
    ]
    available_metrics = [m for m in metrics if m in df_no_errors.columns]

    percent_metrics = {"mAP@0.5", "Precision", "Recall", "F1 score"}

    unit_map = {
        "mAP@0.5": "(%)",
        "Precision": "(%)",
        "Recall": "(%)",
        "F1 score": "(%)",
        "Model size": "(MB)",
        "Average FPS": "(FPS)",
        "Parameters": "(count)"
    }

    def wrap_label(text, width=20):
        return '\n'.join(textwrap.wrap(text, width=width))

    if not available_metrics:
        print("None of the expected metrics are available.")
        return

    # Bar plots with highlighting max for percent_metrics
    for metric in available_metrics:
        plt.figure(figsize=(12, 6))
        labels = df_no_errors["Model"] + " (" + df_no_errors["GPU"] + ")"
        values = pd.to_numeric(df_no_errors[metric], errors='coerce')

        valid = ~values.isna()
        labels = labels[valid]
        values = values[valid]

        bars = plt.bar(labels, values, color="lightgreen", edgecolor="black")
        plt.title(f"{metric} per Model/GPU", fontsize=12)
        plt.ylabel(metric, fontsize=10)
        plt.xlabel("Model (GPU)", fontsize=10)
        plt.xticks(rotation=30, ha='right', fontsize=8)

        # Find max value index for highlighting if metric in percent_metrics
        max_idx = None
        if metric in percent_metrics:
            max_idx = values.idxmax()

        for i, bar in enumerate(bars):
            height = bar.get_height()
            if not pd.isna(height):
                # Highlight max value in yellow and bold
                if max_idx is not None and labels.index[i] == max_idx:
                    color = 'gold'
                    weight = 'bold'
                    size = 9
                else:
                    color = 'black'
                    weight = 'normal'
                    size = 7

                plt.annotate(f'{height:.3f}',
                             xy=(bar.get_x() + bar.get_width() / 2, height),
                             xytext=(0, 3),
                             textcoords="offset points",
                             ha='center', va='bottom',
                             fontsize=size,
                             color=color,
                             fontweight=weight)

        plt.tight_layout()
        plt.savefig(f"chart_{metric.replace(' ', '_')}.png", dpi=300, bbox_inches='tight')
        plt.close()

    # Pie charts with highlighting max value for percent_metrics
    for metric in available_metrics:
        pie_data = df_no_errors.dropna(subset=[metric])
        if pie_data.empty:
            print(f"No valid data for pie chart of '{metric}'. Skipping.")
            continue

        labels = pie_data["Model"] + " (" + pie_data["GPU"] + ")"
        if metric in percent_metrics:
            sizes = pd.to_numeric(pie_data[metric], errors='coerce').fillna(0) * 100
        else:
            sizes = pd.to_numeric(pie_data[metric], errors='coerce').fillna(0)

        if len(labels) > 8:
            top_n = sizes.nlargest(8)
            sizes = top_n
            labels = labels[top_n.index]

        plt.figure(figsize=(11, 11))
        wedges, texts = plt.pie(
            sizes,
            labels=['']*len(labels),
            startangle=90,
            wedgeprops={'edgecolor': 'black'}
        )

        # Find index of max value for highlighting
        max_idx = sizes.idxmax() if metric in percent_metrics else None

        for i, wedge in enumerate(wedges):
            ang = (wedge.theta2 + wedge.theta1) / 2
            x = 0.6 * np.cos(np.deg2rad(ang))
            y = 0.6 * np.sin(np.deg2rad(ang))

            rotation = ang
            if ang > 90 and ang < 270:
                rotation = ang + 180
                ha = 'right'
            else:
                ha = 'left'

            wrapped_label = wrap_label(labels.iloc[i])

            val_str = f"{sizes.iloc[i]:.1f}"
            if metric in percent_metrics:
                val_str += "%"

            # Highlight max value text color gold else white
            is_max = (max_idx is not None and sizes.index[i] == max_idx)
            color = 'gold' if (max_idx is not None and sizes.index[i] == max_idx) else 'white'
            #weight = 'bold' if color == 'gold' else 'normal'
            weight = 'bold'
            fontsize = 11

            plt.text(
                x, y,
                f"{wrapped_label}\n{val_str}",
                ha='center',
                va='center',
                rotation=rotation,
                rotation_mode='anchor',
                fontsize=fontsize,
                color=color,
                fontweight=weight
            )

        unit = unit_map.get(metric, "")
        plt.title(f"{metric} Distribution {unit}", fontsize=14)
        plt.tight_layout()
        plt.savefig(f"pie_{metric.replace(' ', '_')}.png", dpi=300, bbox_inches='tight')
        plt.close()

    print("‚úÖ All charts generated with max values highlighted for specified metrics.")


# -------------------------------
# Main execution block
# -------------------------------
def main():
    """
    Main procedure for multi-GPU inference benchmarking.
    """
    image_dir = "/home/pruebaaitor/Desktop/S_YOLOv8/dataset_balanced/images/test/"
    data_path = "/home/pruebaaitor/Desktop/S_YOLOv8/cfg/datasets/" 
    data=f"{data_path}custom_data5.yaml"
    #image_dir = f"{data_path}custom_data5.yaml"
    model_path = "/home/pruebaaitor/Desktop/S_YOLOv8/model_run_update/newruns/"
    rest_path = "/train/weights/best.pt"
    


    model_paths = {
        "YOLOv8n": f"{model_path}detect_v8{rest_path}",
        "YOLOv10": f"{model_path}detect_v10{rest_path}",
        "YOLOv11": f"{model_path}detect_v11{rest_path}",
        "YOLOv12": f"{model_path}detect_v12{rest_path}",
        "RTDETR": f"{model_path}detect_rtdetr_{rest_path}",
        "SHEPHERD": f"{model_path}detect_CNB_SE_WIoU{rest_path}",
        
    }
    model_paths = {
        "SHEPHERD": f"{model_path}detect_v8CNB{rest_path}",   
    }

    #True : Predict, False : Test set validation 
    predict_mode = False  
    yaml_path = data if not predict_mode else None
    print("üìÅ Loading test images...")
    images, paths = load_images(image_dir, max_images=10, seed=123)

    print("üß† Detecting available GPUs...")
    gpus = get_all_available_gpus.get_available_gpus()
    for idx, name in gpus:
        print(f" - GPU {idx} : {name}")
    try:
        results = []
        for gpu_idx, gpu_name in gpus:
            print(f"\nüöÄ Running inference on GPU {gpu_idx} ({gpu_name})")
            for name, path in model_paths.items():
                print(f" ‚ñ∂Ô∏è Model: {name} for {len(images)} images")
                res = evaluate_model(
                    name, path, images, gpu_idx, gpu_name,
                    predict_mode=predict_mode,
                    yaml_path=yaml_path,
                    save_dir_root="results_predict"
                )
                results.append(res)
    except Exception as e:
        print(f"Warning: Could not retrieve metrics due to error: {e}")

    if not predict_mode:
        #df = pd.DataFrame(results)
        #df.to_csv("inference_multi_gpu_results.csv", index=False)
        #plot_comparison(df)
        print("\n‚úÖ Inference results saved to 'inference_multi_gpu_results.csv'")
        #print(df)

# -------------------------------
# Script entry point
# -------------------------------
if __name__ == "__main__":
    main()
```
